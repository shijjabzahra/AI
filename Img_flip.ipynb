```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/1000.jpeg')
mirrored_img = cv2.flip(img, 1)
print("Original Image:")
cv2_imshow(img)
print("Mirrored Image: ")
cv2_imshow(mirrored_img)


```
